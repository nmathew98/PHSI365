# Equations of motion: Hooke's Law
The complete model of our system without a driving force is:
<div class="alert alert-block alert-warning">
<font color=blue>
\begin{align}
v_n&=\dot{x_n}\\
\dot{v_n}&=-\omega^{2} (2x_n - x_{n+1} - x_{n-1}) - \gamma (2v_n - v_{n+1} - v_{n-1})
\end{align}
</font>
</div>

The end point masses in our system, $x_1$ and $x_N$, are allowed to move freely and only couple to one nearest neighbour, namely, x_2 and x_{N-1}. Then,

$$ 
{\bf X}(t)\equiv \begin{bmatrix}
\mathbf{x}(t)\\
\mathbf{v}(t)\\
\end{bmatrix}
$$

$$
{\bf L}\equiv\begin{bmatrix}
    0_N         & I_N  \\
    -\omega^2 A & -\gamma A
\end{bmatrix},
$$

$$\ale{B_N&=\begin{bmatrix}
    1  & -1 & 0   \\
    -1 &  2 & -1 & \ddots \\
    0  & -1 &  2 & -1 & \ddots\\
       & \ddots  & -1 & 2 & -1 &  0\\
       &         & \ddots & -1 &  2 & -1 \\
       &         &        &  0 & -1 & 1 
\end{bmatrix}
}$$

```julia
 = ;
 = 0;
N = 300; # Number of masses we want to model
alg = AutoVern7(Rodas5()); # Algorithm to use
ti = 0.0;
tp = 0.1;
tf = 600;
Nt = 600;
dt = 1/20;
tspan = (ti, tf); # Timespan
x0 = transpose([1.0 zeros(1, N - 1)]) # Initial values for x(t)
v0 = transpose(zeros(1, N)) # Initial values for v(t)
u0 = [x0; v0] # Initial values for our differential equation, [x(t = 0); v(t = 0)]

# Construct the matrix A
v = ones(N - 1);
boundary_conditions = [1 -1 zeros(1, N - 2)];
middle_conditions = diagm(-1 => -v, 0 => 2*v, 1 => -v);
A = [boundary_conditions; middle_conditions[2:end - 1, :]; reverse(boundary_conditions, dims = 2)];
zero_N = zeros((N, N));
L = [zero_N I; -^(2).*A -.*A] # First three hundred rows set x(t), remaining 300 rows set v(t)
L = sparse(L)

# Our problem
f(du, u, p, t) = mul!(du, L, u)
prob = ODEProblem(f, u0, tspan)
sol = solve(prob, alg=alg, dt=dt, saveat=tp)

@manipulate for i in 1:N
    plot(sol.t, [x[i] for x in sol.u[:]], legend=false);
    title!("Mass " * string(i));
    xlabel!(L"\textrm{Time (s)}");
    ylabel!(L"\textrm{Displacement (m)}");
    yticks!([-0.2:0.1:0.2;]);
    ylims!(-0.2, 0.2);
    xlims!(0, Inf);
    end
```
