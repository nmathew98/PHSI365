# A. Vertical motion

1. The equation expressing the condition that $y(t) = h$ is $y(t) = - \frac{1}{2}gt^2 + v_0t  = h$. This condition in the form $F(t)=0$ is $F(t) =  - \frac{1}{2}gt^2 + v_0t - h = 0$ 


2. The roots of the equation are: $\frac{-v_0 \ \pm \ \sqrt{v_0^2 - 2gh}}{-g}$. The physical constraints are that the initial velocity has to be greater than or equal to $\sqrt{2gh}$.


3. The following variables are defined in the code cell below.
```julia
g = 9.8 ± 0.014
m = 23 ± 0.0037
v_0 = 21 ± 0.1
h = 20 ± 0.4
y_0 = 0.0 ± 0.018
```


4. The roots are at approximately 1.5 and 2.8. The following code is used to plot $y(t)$ with uncertainties over time. On the same plot, a line of constant height $h$ was also plotted including its error.
```julia
# Plot y(t) against time
time = collect(0:0.5:10) # Generate a vector with values from 0 to 10 in increments of 0.5
y(t) = -1/2*g*t^2 + v_0*t # Equation for y(t)
plot(y, [t ± 0.1 for t in time], size = (1200, 800)) # Plot y(t) against time
plot!([h ± 0.4], seriestype="hline") # Line of constant height h
```


5. The roots of $y(t)$ found using the quadratic formula including uncertainties are 1.429 ± 0.021 and 2.857 ± 0.042. The roots found using the quadratic formula are 1.429 ± 0.021 and 2.857 ± 0.042. The code below was used.
```julia
discriminant(a, b, c) = b^2 - 4*a*c # Find the discriminant of the quadratic equation
findRoots(a, b, c) = (1/(2*a))*((-b)*[1, 1] + sqrt(discriminant(a, b, c))*[1, -1]) # Find the roots of the quadratic equation
```


6. The equation of the curve in the form $F(t) = 0$ is $F(t) =  - \frac{1}{2}gt^2 + v_0t - h = 0$. The code below was used to plot $F(t)$ against time.
```julia
F(t) = -1/2*g*t^2 + v_0*t - h # Equation for F(t) = 0
plot!(F, [t ± 0.1 for t in time], size = (1200, 800), legend=false) # Plot F(t) = 0
```


7. The solutions to $F(t) = 0$ found numerically are 1.429 ± 0.061 and 2.857 ± 0.071. The code below was used to find the solutions.

````julia
# Variable declarations
g = 9.8 ± 0.014;
m = 23 ± 0.0037;
v_0 = 21 ± 0.1;
h = 20 ± 0.4;
y_0 = 0.0 ± 0.018;
````

````julia
# Code for A(4)
time = collect(0:0.5:10)
y(t) = -1/2*g*t^2 + v_0*t
plot(y, [t ± 0.1 for t in time], size = (1200, 800), legend=true, title = "Graph for A(4)", label=L"y(t) = -0.5(gt^2) + v_0t")
plot!([h ± 0.4], seriestype="hline", label=L"y(t) = 20") # Horizontal line y(t) = h = 20

# Plots attributes
xlabel!("Time, t (s)")
ylabel!("Height, h (m)")
xlims!(0, 7)
ylims!(-20, 60)
````
