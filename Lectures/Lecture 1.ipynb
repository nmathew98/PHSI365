## Physics

In 2017 the Nobel Prize in Physics awarded just one year after LIGO's astonishing observations of [gravity waves emitted by two black holes merging](https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.116.061102). 

As physicists, we are curious problem solvers, and our basic trade is insight. Not just any kind of insight. We aim for a complete, rigorous, and quantitative understanding of the governing principles of nature - at least, that part of nature that can be understood mathematically. 

In addition to detailed observations of nature, our work is thus fundamentally mathematical, and oftentimes computational: a process of either generating mathematical models and their numerical predictions, or testing predictions against physical observations. 

- [What computational physics is really about](https://www.wired.com/2015/11/what-computational-physics-is-really-about/)

## Insight
Far from being automatic insight generators, computers are just simple number crunchers (notwithstanding recent advances in congnitive ability, deep neural networks such as AlphaGo).

In order to do computational physics well, it is essential to 

- Have a solid understanding of the physics involved


- Have a keen eye for debugging, "sanity checking" and 


- Take a sober, skeptical view of your results

By debugged, sanity checking, and testing, we build confidence in the validity of our results, and their interpretation.

__Fun fact:__ Which pioneering scientist carried out the first computer experiments? 

Studies of nonlinear dynamics in a chian of oscillators were carried out on the [MANIAC-1](https://en.wikipedia.org/wiki/MANIAC_I) (Los Alamos) in 1953 by [Mary Tsingou](https://en.wikipedia.org/wiki/Fermi%E2%80%93Pasta%E2%80%93Ulam%E2%80%93Tsingou_problem).

# What is Julia?

> _"We want a language that’s open source, with a liberal license. We want the speed of C with the dynamism of Ruby. We want a language that’s homoiconic, with true macros like Lisp, but with obvious, familiar mathematical notation like Matlab. We want something as usable for general programming as Python, as easy for statistics as R, as natural for string processing as Perl, as powerful for linear algebra as Matlab, as good at glueing programs together as the shell. Something that is dirt simple to learn, yet keeps the most serious hackers happy. We want it interactive and we want it compiled.

## How to read and test a package

I want to emphasize that while the open-source world may look a little disorganized, don't be fooled: GitHub style code development is the leading industry standard, and leads to very reliable processes for development, testing and bugfixing. 

This brings me to my point: 

- all development is ___test-driven___, which means that there should be an extensive set of tests that pacakge code is subjected to before any updates go live. You can always test a package to see if it is passing its own tests. 


- Furthermore, if documentation is a bit light, you can read the test code to get some understanding of what the code is supposed to do. 


- You can also always read the source code since it is freely available. This is much easier than you might think: yay julia!
